```python
import pandas as pd
finale_car=pd.read_csv("new_cars.csv")
finale_car.drop(columns="Unnamed: 0",inplace=True)
car_colors=pd.read_csv('Colors.csv',header=None,names=['Color'])
```

```python
colu=pd.unique(finale_car["Extrior"])
```

```python
#giving each color a unique number
col=car_colors.Color
Colour={}
colo=col.tolist()
i=0
while i<=len(colo)-1:
    j=colo[i]
    Colour[j]=i
    i=i+1
```

```python
# giving each brand a unique number
brand=pd.unique(finale_car["Brand"])
brands={}
brand_M=brand.tolist()
i=0
while i<=len(brand_M)-1:
    j=brand_M[i]
    brands[j]=i
    i=i+1
```

```python
brand_m=list(finale_car["Brand"])
```

```python
#conversion of brand to numeric  
list_brans = []
i=0
while i<len(brand_m):
    k=brand_m[i]
    numd=brands[k]
    list_brans.append(numd)
    i=i+1   
finale_car["Bran_N"]=list_brans
```

```python
# checking if a color is in the name if not adds it as other and converts them to factors 
Ext=finale_car.Extrior
list_colu= []
list_col = []
i=0
for Exto in Ext:
    Ex=Exto.lower()
    for Co in col:      
        cole=Co.lower()
        if cole in Ex:
            list_col.append(Co)
            break
    else:
        list_col.append("Others")
while i<len(list_col):
    k=list_col[i]
    numd=Colour[k]
    list_colu.append(numd)
    i=i+1   
finale_car["Color"]=list_col
finale_car["Color_num"]=list_colu
```

```python
finale_car.to_csv("finale_car.csv",header=True)
```
